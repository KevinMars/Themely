```python
from indicoio import sentiment, political, text_tags
from nltk import pos_tag
from nltk.corpus import wordnet, stopwords
from nltk.tokenize import word_tokenize,WordPunctTokenizer
from operator import itemgetter
from pymongo import MongoClient
from pprint import pprint
import re, string
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN, MiniBatchKMeans
from textblob import TextBlob, sentiments

nas = MongoClient().dsbs.nas
```

```python
song_query = nas.find()
songs = []
for i in song_query:
    songs.append(i['lyrics']+i['annotations'])
```

```python
songs = [song.lower().replace("'s","").replace("'","").replace(u"’","").replace(u"—","").replace("/","")\
        .replace("[intro"," ").replace('[hook'," ").replace('[verse 1'," ").replace('[verse 2'," ")\
        .replace('[verse 3'," ").replace('[verse 4'," ") for song in songs]
```

```python
songs = [re.sub('http\S+', "", song) for song in songs]
```

```python
pattern = re.compile('[\W_]+')
songs = [pattern.sub(' ', song) for song in songs]
```

```python
songs = [" ".join([w for w in set(TextBlob(song).words.lower())]) for song in songs]
```

```python
for idx, song in enumerate(songs):
    for word in TextBlob(song).words:
        if len(wordnet.synsets(str(word))) == 0:
            songs[idx] = songs[idx].replace(str(word), "")
            
```

```python
vectorizer = TfidfVectorizer(stop_words='english')

song_vector = vectorizer.fit_transform(songs)
```

```python
from gensim.matutils import Sparse2Corpus
corpus = Sparse2Corpus(song_vector.transpose())
```

```python
from gensim.models.ldamodel import LdaModel
from gensim.corpora import Dictionary

lda = LdaModel(corpus=corpus, num_topics=30, passes=30, iterations=300)
```
